```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib
import pylab as P
import matplotlib.pyplot as plt

Juni=pd.read_csv('data/November.csv')
Juni_BASE=Juni.loc[Juni['ANLASS'] == 'V1']
Juni_EOS=Juni.loc[Juni['ANLASS'] != 'V1']
print(Juni)
#display(Juni_EOS)
Juni_UQEEMED=Juni['UQEEMED'].dropna()
Juni_UQEEMED_BASE=Juni_BASE['UQEEMED'].dropna()
Juni_UQEEMED_EOS=Juni_EOS['UQEEMED'].dropna()
#display(Juni_UQEEMED_BASE)
Juni_DATE=Juni['UDATUM'].dropna()

#START="20120101"
#MONTHS_SINCE_START=Juni_DATE-Start in months

#dateformat.parse('%Y%m%d', START)


P.figure()
plt.hist(Juni_UQEEMED_BASE, rwidth=0.8)
plt.title("Juni Baseline+EoS")
plt.xlabel("E/e' medial")
plt.ylabel("Frequency")
plt.hist(Juni_UQEEMED_EOS, rwidth=0.8,color='orange')
plt.xlabel("E/e' medial")
plt.ylabel("Frequency")

P.figure()
plt.hist(Juni_UQEEMED_BASE, rwidth=0.8)
plt.title("Juni Baseline")
plt.xlabel("E/e' medial")
plt.ylabel("Frequency")


P.figure()
plt.hist(Juni_UQEEMED_EOS, rwidth=0.8,color='orange')
plt.title("Juni EoS")
plt.xlabel("E/e' medial")
plt.ylabel("Frequency")



P.figure()
plt.hist(Juni_DATE, rwidth=0.8,color='orange')
plt.title("Datum")
plt.xlabel("Month")
plt.ylabel("Frequency")



```
